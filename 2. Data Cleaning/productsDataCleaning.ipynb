```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Products.csv')
data.head()
```

```python
data.info()
```

```python
# checking for nulls
data.isnull().sum()
```

```python
# convert unit_cost_usd and unit_price_usd to float
data['Unit Cost USD'] = data['Unit Cost USD'].str.replace('[$,]', '', regex=True).str.strip().astype(float)
data['Unit Price USD'] = data['Unit Price USD'].str.replace('[$,]', '', regex=True).str.strip().astype(float)
```
